```python
import csv
from datetime import datetime

expenses = []
budget = 0
filename = "expenses.csv"


# add expense
def add_expense():

    while True:
        date = input("Date (YYYY-MM-DD): ")
        try:
            date = datetime.strptime(date,"%Y-%m-%d").date()
            break
        except ValueError:
            print('Invalid date format. Please enter in YYYY-MM-DD format')

    category= input("Category (Food, Travel etc): ")
    try:
        amount = float(input("Amount: Rs. "))
    except:
        print("Please Enter a valid Number")
    description = input("Description: ")

    expense = {
        'date':date,
        'category':category,
        'amount': amount,
        'description':description
    }
    expenses.append(expense)
    print("Expense added !")

# view_expenses
def view_expenses():
    print("\n === Your Expenses ===")

    if not expenses:
        print("No expenses to show")
        return

    for e in expenses:
        print(f"{e['date']} | {e['category']} | Rs. {e['amount']} | {e['description']}")

# track_budget
def track_budget():
    print("\n === Track budget ===")

    try:
        budget = float(input("Enter your monthly budget: Rs. "))
    except ValueError:
        print("Invalid Input")
        return

    total_expenses = sum(e['amount'] for e in expenses)
    left = budget - total_expenses

    if left < 0:
        print("You have exceeded your budget!")
    else:
        print(f"You have {left} left for the month")

# save_expenses
def save_expenses():
    with open(filename, "w", newline="") as file:
        fieldnames = ['date', 'category', 'amount', 'description']
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        writer.writeheader()
        for exp in expenses:
            writer.writerow(exp)
    print("Expenses saved to file.")


# load_expenses
def load_expenses():
    try:
        with open(filename,"r") as file:
            reader = csv.DictReader(file)
            for row in reader:
                row['amount'] = float(row["amount"])
                expenses.append(row)
        print("Expnses loaded from file.")

    except FileNotFoundError:
        print("No saved file found.")

# interactive_menu
def interactive_menu():
    print("\n=== Menu ===")
    print("1. Add Expense")
    print("2. View Expenses")
    print("3. Set and track the budget")
    print("4. Save Expenses")
    print("5. Exit")


# Main function
def main():
    load_expenses()

    while True:
        interactive_menu()

        choice = input("Choose 1-5: ")

        if choice == '1':
            add_expense()
        elif choice == '2':
            view_expenses()
        elif choice == '3':
            track_budget()
        elif choice == '4':
            save_expenses()
        elif choice == '5':
            save_expenses()
            break
        else:
            print("Pease Enter A Number from 1 To 5")

if __name__ == "__main__":
    main()
```
